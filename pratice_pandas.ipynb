```python
def hi(percentage):
    if percentage>90:
        return "Garde A+"
    elif percentage>80:
        return "Grade A"
    elif percentage>70:
        return "Grade B"
    elif percentage>60:
        return "Grade C"
    elif percentage>50:
        return "Grade D"
    elif percentage>35:
        return "Passed"
    else:
        return "Failed"
```

```python
result["Grade"]=result.Percentage.apply(hi)
```
